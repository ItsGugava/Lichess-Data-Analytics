```python
import matplotlib.pyplot as plt
import numpy as np

moves = data['moves']


first_moves = np.array([move.split()[0] for move in moves])
labels, sizes = np.unique(first_moves, return_counts=True)


plt.figure(figsize=(8, 6))
plt.bar(labels, sizes)

for i in range(len(labels)):
    plt.text(i, sizes[i] + 0.5, str(sizes[i]), ha='center')

plt.title('First Move Frequency')
plt.xlabel('Moves')
plt.ylabel('Amount')


plt.show()



```

```python
white_higher = data[data['white_rating'] > data['black_rating']]
black_higher = data[data['black_rating'] > data['white_rating']]

white_win_white_higher = (white_higher['winner'] == 'white').sum()
black_win_white_higher = (white_higher['winner'] == 'black').sum()
draw_white_higher = (white_higher['winner'] == 'draw').sum()

white_win_black_higher = (black_higher['winner'] == 'white').sum()
black_win_black_higher = (black_higher['winner'] == 'black').sum()
draw_black_higher = (black_higher['winner'] == 'draw').sum()


labels_white_higher = ['White Win', 'Black Win', 'Draw']
sizes_white_higher = [white_win_white_higher, black_win_white_higher, draw_white_higher]


labels_black_higher = ['White Win', 'Black Win', 'Draw']
sizes_black_higher = [white_win_black_higher, black_win_black_higher, draw_black_higher]


plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.pie(sizes_white_higher, labels=labels_white_higher, autopct='%1.1f%%', startangle=140)
plt.title('White had higher rating')


plt.subplot(1, 2, 2)
plt.pie(sizes_black_higher, labels=labels_black_higher, autopct='%1.1f%%', startangle=140)
plt.title('Black had higher rating')

plt.tight_layout()
plt.show()
```
